# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module o Matplotlib.
Matplotlib is a popular plotting library in Python used for creating static, animated, and interactive visualizations. It provides a comprehensive set of functions and classes for generating various types of plots and charts.

Matplotlib is used for data visualization and is beneficial in several ways:

1. Creating Publication-Quality Plots: Matplotlib allows you to create high-quality plots with customizable features, such as labels, titles, legends, colors, and styles, to make the plots visually appealing and suitable for publications or presentations.

2. Exploring Data: Matplotlib enables data analysts and scientists to explore and understand data by visualizing it in different ways. It helps in identifying patterns, trends, outliers, and relationships between variables.

3. Communication and Presentation: Plots created with Matplotlib can effectively communicate insights and findings to a wider audience. Visualizations are often easier to understand and interpret than raw data, making them useful for presentations, reports, and storytelling.

4. Integration with Scientific Libraries: Matplotlib integrates seamlessly with other scientific libraries in the Python ecosystem, such as NumPy, Pandas, and SciPy. It complements these libraries by providing visualization capabilities for data analysis and scientific computing workflows.

5. Flexibility and Customizability: Matplotlib offers a wide range of plot types and customization options, allowing you to create a variety of visualizations tailored to your specific needs. It provides flexibility in terms of plot types, layout arrangements, color schemes, annotations, and interactive features.

Five commonly used plots that can be created using the `pyplot` module of Matplotlib are:

1. Line Plot: Displays data points connected by straight lines, commonly used to visualize trends over time or continuous data.

2. Scatter Plot: Shows the relationship between two variables by plotting individual data points as dots. It helps identify patterns or correlations between variables.

3. Bar Plot: Represents data using rectangular bars of different heights, often used to compare categories or show the distribution of categorical data.

4. Histogram: Illustrates the distribution of continuous or discrete data by dividing it into bins and displaying the frequency or count of observations in each bin.

5. Pie Chart: Represents data as slices of a circular pie, where the size of each slice corresponds to the proportion of each category in the dataset. It is suitable for displaying proportions or percentages of a whole.


# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter Plot shows the relationship between two variables by plotting individual data points as dots. It helps identify patterns or correlations between variables.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))

plt.scatter(x,y,c='g')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter plot')

plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple subplots within a single figure. It provides a way to arrange and organize multiple plots or axes in a grid-like structure, allowing for more complex and informative visualizations.

```python
fig = plt.figure()

x=np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig,ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].plot(x,y1,c='b')
ax[0,0].set_title('Line 1')

ax[0,1].plot(x,y2,c='r')
ax[0,1].set_title('Line 2')

ax[1,0].plot(x,y3,c='g')
ax[1,0].set_title('Line 3')

ax[1,1].plot(x,y4,c='y')
ax[1,1].set_title('Line 4')


plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart or bar graph, is a type of visualization that uses rectangular bars to represent categorical data. The length or height of each bar corresponds to the value or frequency of the category it represents. Bar plots are commonly used to compare different categories or display the distribution of categorical data.

Bar plots are used for several purposes:

1. Comparing Categories: Bar plots provide a visual representation of the values or frequencies associated with different categories. They allow for easy comparison between categories by visually comparing the lengths or heights of the bars. This makes bar plots useful for illustrating contrasts or relationships between different groups or variables.

2. Displaying Categorical Data: Bar plots are especially suitable for displaying categorical data, where each category represents a distinct group or class. They help in visualizing the distribution or proportions of different categories within a dataset.

3. Presenting Rankings or Ordering: Bar plots can be used to present rankings or ordering of categories based on their values or frequencies. By arranging the bars in ascending or descending order, you can highlight the relative magnitudes or importance of the categories.

4. Exploring Patterns or Trends: Bar plots can reveal patterns or trends within categorical data. By observing the differences in bar lengths or heights, you can identify variations, outliers, or patterns in the data that may provide insights into underlying patterns or relationships.

5. Communicating Findings: Bar plots are effective in communicating findings or results to a broad audience. They provide a clear and concise visual representation of data, making it easier for others to understand and interpret the information being presented.

Bar plots can be customized to enhance their visual impact, such as adding labels, titles, colors, and additional annotations. They are widely used in various fields, including statistics, business, social sciences, marketing, and data analysis, to visualize categorical data and make informed comparisons between different categories or groups.


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

fig,ax = plt.subplots(1,2,figsize=(12,5))

ax[0].bar(company,profit)
ax[0].set_title('Bar plot')

ax[1].barh(company,profit,color='g')
ax[1].set_title('Horizontal Bar Plot')



plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, is a statistical visualization that provides a summary of the distribution of a continuous variable. It displays the median, quartiles, and potential outliers of the data in a concise and informative manner. Box plots are used to gain insights into the central tendency, spread, and skewness of the data.

```python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)



plt.boxplot([box1,box2])
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Values')
plt.title('Box Plot')

plt.show()
```
