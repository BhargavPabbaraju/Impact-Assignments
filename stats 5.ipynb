# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

**Estimation Statistics**:

Estimation in statistics refers to the process of making inferences about population parameters based on sample data. It involves using sample statistics to estimate the values of unknown population parameters. Estimation is a crucial aspect of statistical analysis because it allows us to draw conclusions about a population when it is not feasible or practical to measure the entire population.

There are two main types of estimation:

1. **Point Estimate**: A point estimate is a single value that is used to estimate an unknown population parameter. It is calculated using a sample statistic that is expected to be close to the true population parameter. For example, the sample mean is a point estimate of the population mean, and the sample proportion is a point estimate of the population proportion.

2. **Interval Estimate**: An interval estimate is a range of values within which the true population parameter is likely to lie, along with an associated level of confidence. It provides a measure of uncertainty associated with the point estimate. The confidence interval is a common form of interval estimate, and it is calculated based on the sample statistic and the level of confidence chosen by the analyst.

**Point Estimate**:

A point estimate is a single numerical value that serves as the best guess or approximation of an unknown population parameter based on sample data. Point estimates are calculated using sample statistics and are typically denoted by specific symbols.

For example, if we want to estimate the population mean $\mu$, we can calculate the sample mean $\bar{x}$ and use it as the point estimate:

$$ \text{Point Estimate of } \mu: \bar{x} $$

Similarly, if we want to estimate the population proportion (\(p\)), we can calculate the sample proportion $\hat{p}$ and use it as the point estimate:

$$\text{Point Estimate of } p: \hat{p} $$

**Interval Estimate**:

An interval estimate provides a range of values within which the true population parameter is likely to lie, along with an associated level of confidence. It takes into account the variability in the sample data and provides a measure of uncertainty.

For example, we can calculate a 95% confidence interval for the population mean (\(\mu\)) using the sample mean $\bar{x}$ and the standard error of the mean $SE_{\bar{x}}$:

$$ \text{Confidence Interval for } \mu: \bar{x} \pm 1.96 \times SE_{\bar{x}} $$

This interval estimate means that we are 95% confident that the true population mean falls within the range defined by $\bar{x} \pm 1.96 \times SE_{\bar{x}}$.



# Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.


```python
def get_pop_mean(sample_mean,sample_std,sample_size):
    standard_error = sample_std / (sample_size**(1/2))
    
    confidence_level = 1.96
    margin_of_error = confidence_level * standard_error
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    pop_mean = (lower_bound + upper_bound) / 2
    
    return pop_mean


get_pop_mean(25.4,3.6,50)
```

# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

**Hypothesis testing** is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses: the null hypothesis $H_0$ and the alternative hypothesis $H_1$ or $H_a$. These hypotheses are used to assess the validity of certain claims or assumptions about the population.

The process of hypothesis testing involves the following steps:

1. **Formulate Hypotheses**: The null hypothesis $H_0$ represents the default assumption, often stating that there is no significant effect, no difference, or no relationship between variables. The alternative hypothesis $H_1$ or $H_a$ represents the claim or research question that contradicts the null hypothesis, suggesting that there is a significant effect, difference, or relationship.

2. **Select a Test Statistic**: Based on the data and the hypotheses, a test statistic is selected to quantify the evidence against the null hypothesis. The choice of test statistic depends on the type of data and the research question.

3. **Set the Significance Level**: The significance level $\alpha$ is the probability of rejecting the null hypothesis when it is true (Type I error). It represents the threshold for the level of evidence required to reject the null hypothesis. Commonly used significance levels are 0.05 (5%) and 0.01 (1%).

4. **Calculate the p-value**: The p-value is the probability of observing the data, or more extreme data, under the assumption that the null hypothesis is true. It quantifies the strength of the evidence against the null hypothesis. If the p-value is smaller than the significance level $\alpha$, the null hypothesis is rejected in favor of the alternative hypothesis.

5. **Make a Decision**: Based on the p-value and the chosen significance level, the decision is made to either reject the null hypothesis or fail to reject it. If the p-value is less than or equal to $\alpha$, the null hypothesis is rejected, indicating that there is evidence to support the alternative hypothesis. Otherwise, if the p-value is greater than $\alpha$, the null hypothesis is not rejected due to lack of evidence.

**Importance of Hypothesis Testing**:

Hypothesis testing is a fundamental and crucial process in statistics and scientific research for several reasons:

1. **Scientific Inquiry**: Hypothesis testing enables researchers to systematically evaluate and test their theories, ideas, or hypotheses using empirical evidence.

2. **Decision Making**: Hypothesis testing provides a structured framework for making decisions based on data, allowing us to draw conclusions about population parameters.

3. **Inference**: It allows us to generalize findings from a sample to a larger population, provided the sample is representative and the assumptions of the test are met.

4. **Validity of Results**: Hypothesis testing helps assess the validity of claims or assumptions by subjecting them to rigorous statistical evaluation.

5. **Comparison of Groups**: Hypothesis testing is used to compare groups or conditions, such as comparing treatment effects in experiments or comparing characteristics of different populations.

6. **Statistical Significance**: It provides a measure of statistical significance, indicating whether an observed effect is likely due to chance or a real effect.

7. **Objectivity**: Hypothesis testing provides an objective and quantitative approach to decision-making, reducing the influence of subjective opinions.

# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.



```python
import numpy as np
from scipy.stats import ttest_ind

def hypothesis_test_two_means(sample1, sample2, alpha=0.05):
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)

    null_hypothesis_rejected = p_value < alpha

    return null_hypothesis_rejected, p_value


sample1 = np.array([10, 12, 15, 18, 20])
sample2 = np.array([8, 11, 14, 17, 19])

alpha = 0.05  

result, p_value = hypothesis_test_two_means(sample1, sample2, alpha)

if result:
    print("Null hypothesis rejected. There is a significant difference in means.")
else:
    print("Null hypothesis not rejected. There is no significant difference in means.")

print("P-value:", p_value)
```

# Q6: What is a null and alternative hypothesis? Give some examples.

**Null Hypothesis**:

The null hypothesis $H_0$ is a statement of no effect or no difference between groups. It is a default assumption that there is no significant relationship or effect in the population being studied. Researchers often begin by assuming that the null hypothesis is true and then conduct statistical tests to see if there is enough evidence to reject it in favor of an alternative hypothesis.

In statistical hypothesis testing, the null hypothesis is typically denoted with an equal sign or a less than or equal to sign (depending on the test), indicating that there is no significant difference or effect.

**Examples of Null Hypotheses**:
1. There is no difference in the average test scores between two groups.
   $$H_0: \mu_1 = \mu_2$$

2. A new drug has no effect on patients' blood pressure.
   $$H_0: \text{The mean blood pressure after taking the drug} = \text{The mean blood pressure before taking the drug}$$

3. There is no relationship between age and job performance.
   $$H_0: \text{There is no correlation between age and job performance}$$

**Alternative Hypothesis**:

The alternative hypothesis $H_1$ or $H_a$ is a statement that contradicts the null hypothesis. It represents the claim or research question for which the researcher seeks evidence. The alternative hypothesis suggests that there is a significant relationship or effect in the population being studied, different from what the null hypothesis assumes.

The alternative hypothesis can take different forms, depending on the research question, and it is the hypothesis that researchers aim to support with their data and analysis.

**Examples of Alternative Hypotheses**:
1. The average test scores of Group 1 are significantly higher than the average test scores of Group 2.
   $$H_1: \mu_1 > \mu_2$$

2. The new drug reduces patients' blood pressure compared to before taking the drug.
   $$H_1: \text{The mean blood pressure after taking the drug} < \text{The mean blood pressure before taking the drug}$$

3. There is a positive correlation between age and job performance.
   $$H_1: \text{There is a positive correlation between age and job performance}$$

In hypothesis testing, researchers collect data and use statistical tests to evaluate the evidence against the null hypothesis. If the evidence is strong enough, the null hypothesis may be rejected in favor of the alternative hypothesis, indicating that there is a significant relationship or effect in the population. If the evidence is not strong enough, the null hypothesis is not rejected, and no significant relationship or effect is concluded based on the data.

# Q7: Write down the steps involved in hypothesis testing.

Hypothesis testing involves a systematic process to make decisions or draw conclusions about a population based on sample data. Here are the general steps involved in hypothesis testing:

1. **Formulate Hypotheses**: Clearly state the null hypothesis (\(H_0\)) and the alternative hypothesis (\(H_1\) or \(H_a\)) based on the research question or claim. The null hypothesis represents the default assumption of no effect or no difference, while the alternative hypothesis represents the claim or research question that contradicts the null hypothesis.

2. **Select a Test Statistic**: Choose an appropriate test statistic that will be used to evaluate the evidence against the null hypothesis. The choice of test statistic depends on the type of data, the research question, and the assumptions of the test.

3. **Set the Significance Level**: Determine the significance level (\(\alpha\)) for the test, which represents the probability of rejecting the null hypothesis when it is true. Commonly used significance levels are 0.05 (5%) and 0.01 (1%). The significance level determines the threshold for the level of evidence required to reject the null hypothesis.

4. **Collect Data and Calculate the Test Statistic**: Collect data from a sample and calculate the value of the chosen test statistic using the sample data. For example, if comparing means between two groups, the test statistic could be a t-statistic or z-statistic.

5. **Calculate the p-value**: The p-value is the probability of observing the data, or more extreme data, under the assumption that the null hypothesis is true. Calculate the p-value corresponding to the calculated test statistic.

6. **Make a Decision**: Compare the p-value to the significance level (\(\alpha\)). If the p-value is less than or equal to \(\alpha\), the evidence is considered significant, and the null hypothesis is rejected in favor of the alternative hypothesis. If the p-value is greater than \(\alpha\), there is not enough evidence to reject the null hypothesis.

7. **Interpret the Results**: If the null hypothesis is rejected, interpret the results in the context of the research question or claim. If the null hypothesis is not rejected, state that there is insufficient evidence to support the alternative hypothesis.

8. **Conclusion**: Summarize the findings of the hypothesis test, including the decision made and the interpretation of the results. Be clear and concise in reporting the conclusions.

9. **Consider Limitations**: Discuss any limitations or assumptions of the test and acknowledge any potential sources of error or bias in the analysis.

10. **Repeat if Necessary**: If the results are inconclusive or new information becomes available, the hypothesis testing process can be repeated with different data or with modifications to the hypotheses or assumptions.



# Q8. Define p-value and explain its significance in hypothesis testing.

The **p-value** (probability value) is a fundamental concept in hypothesis testing. It is the probability of obtaining the observed results (or more extreme results) when the null hypothesis is true. In other words, the p-value measures the strength of the evidence against the null hypothesis based on the sample data.

**Significance of the p-value in Hypothesis Testing**:

The p-value plays a critical role in hypothesis testing and is used to make decisions about whether to reject or fail to reject the null hypothesis. Here's how the p-value is interpreted in hypothesis testing:

1. **p-value â‰¤ Significance Level $\alpha$**: If the calculated p-value is less than or equal to the chosen significance level $\alpha$, typically 0.05 (5%) or 0.01 (1%), then the evidence is considered **statistically significant**. In this case, we reject the null hypothesis and support the alternative hypothesis. The smaller the p-value, the stronger the evidence against the null hypothesis.

2. **p-value > Significance Level $\alpha$**: If the calculated p-value is greater than the chosen significance level $\alpha$, then the evidence is considered **not statistically significant**. In this case, we fail to reject the null hypothesis, meaning there is insufficient evidence to support the alternative hypothesis. The larger the p-value, the weaker the evidence against the null hypothesis.

3. **Interpretation**: When reporting the results of a hypothesis test, it is common to provide the p-value along with the conclusion. For example, if the p-value is 0.03 and the significance level is 0.05, the conclusion might be: "The null hypothesis is rejected at the 5% significance level, as the p-value (0.03) is less than 0.05."

4. **No Direct Proof of the Null Hypothesis**: It's important to note that the p-value does not provide evidence in support of the null hypothesis. It only quantifies the evidence against the null hypothesis. If the p-value is not statistically significant, it does not mean the null hypothesis is true; it simply means that there is not enough evidence to reject it based on the given data.

5. **Not a Measure of Effect Size**: The p-value is not a measure of the magnitude of the effect or the practical significance of the results. It only addresses the question of whether the observed results are likely to occur by chance alone when the null hypothesis is true.

# Q11: What is Studentâ€™s t distribution? When to use the t-Distribution.

**Student's t-distribution** (or simply t-distribution) is a probability distribution that arises in statistical inference when the sample size is small and the population standard deviation is unknown. It is used to make inferences about population parameters (such as the mean) based on sample data when the underlying population follows a normal distribution.

The t-distribution is similar in shape to the standard normal (Z) distribution, but it has heavier tails, making it more appropriate for small sample sizes. As the sample size increases, the t-distribution approaches the standard normal distribution. The shape of the t-distribution is determined by the degrees of freedom $df$, which is related to the sample size.

**When to use the t-Distribution**:

The t-distribution is used in situations where the following conditions are met:

1. **Small Sample Size**: When the sample size is small (typically less than 30) and the underlying population follows a normal distribution, the t-distribution is used for making inferences about the population parameters. For larger sample sizes (e.g., greater than 30), the t-distribution closely approximates the standard normal distribution, and the Z-test can be used.

2. **Unknown Population Standard Deviation**: When the population standard deviation is unknown, the t-distribution is used instead of the Z-distribution, which requires the known population standard deviation.

3. **Sampling Distribution of the Mean**: The t-distribution is commonly used for hypothesis testing and constructing confidence intervals for the population mean $\mu$. For example, when comparing the means of two independent samples or conducting a one-sample t-test.

4. **Outliers and Skewness**: The t-distribution is robust to mild departures from normality, making it suitable for cases where the data may have some skewness or outliers, but the sample size is small.

5. **When Population Standard Deviation is Unavailable**: In many real-world situations, the population standard deviation is unknown or impractical to estimate accurately. The t-distribution provides a more appropriate alternative for such scenarios.



# Q12: What is t-statistic? State the formula for t-statistic.
The **t-statistic** is a test statistic used in hypothesis testing to assess whether there is a significant difference between the means of two groups (independent samples) or between the sample mean and a known or hypothesized population mean (one-sample case). It measures the difference between sample means relative to the variability in the data.

**Formula for t-statistic**:

For a **one-sample t-test**, the formula for calculating the t-statistic is:

$$ t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}} $$

Where:
- $ \bar{x} $ is the sample mean.
- $ \mu $ is the population mean (or the hypothesized mean under the null hypothesis).
- $ s $ is the sample standard deviation.
- $ n $ is the sample size.

For an **independent two-sample t-test**, the formula for calculating the t-statistic to compare the means of two groups is:

$$ t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}} $$

Where:
- $ \bar{x}_1 $ is the mean of group 1.
- $ \bar{x}_2 $ is the mean of group 2.
- $s_1 $ is the standard deviation of group 1.
- $ s_2 $ is the standard deviation of group 2.
- $ n_1 $ is the sample size of group 1.
- $ n_2 $ is the sample size of group 2.

The t-statistic measures how many standard errors the sample mean is away from the hypothesized population mean (one-sample case) or how many standard errors the difference between the sample means is from zero (two-sample case). The larger the absolute value of the t-statistic, the greater the evidence against the null hypothesis.

In hypothesis testing, the t-statistic is compared to a critical value from the t-distribution (or its corresponding p-value is used) to determine whether to reject the null hypothesis or not. If the t-statistic is sufficiently large (or the p-value is sufficiently small), the null hypothesis is rejected, indicating that there is a significant difference between the means of the groups or between the sample mean and the hypothesized population mean. Otherwise, if the t-statistic is not large enough (or the p-value is not small enough), the null hypothesis is not rejected, indicating no significant difference.
